# Simulation results from *Director Field Response of Liquid Crystals in Spatially Varying Magnetic Fields with Antagonistic Anchoring Conditions*

We have developed an experimental strategy for the spatially resolved control of LC director fields by leveraging their anisotropic response to magnetic forces. Here, patterned magnetic fields generated from ferromagnetic anchoring objects embedded into nematic liquid crystals (NLCs) selectively manipulate the LC director's orientation. These micro-structured ferromagnets introduce spatial variations in field strength and direction, thus disrupting the uniformity of a background magnetic field.

We explore geometries of magnetic microstructures and the director field configurations that arise from their respectively formed magnetic field patterns. Here we recreate the results using simulations of the spatial variation of the magnetic field and LC director field.

In tandem, we investigate the synergistic and competitive interplay between magnetic and elastic effects on the orientational ordering of NLCs. Using single-color light and filter, we probe the dynamic response of the liquid crystal to temperature-induced elasticity changes and changes in background field strength. We simulate the movement of periodic fringes near the edges of ferromagnetic structures.

## Experimental system

The NLC cell comprises two glass slides treated to achieve homeotropic anchoring conditions on their surfaces. The ferromagnetic structures are fabricated from 50 Î¼m thick cobalt foils and are commissioned to a custom geometry on a mm scale. The desired cobalt structure is then sandwiched between two glass slides bonded together using optical adhesives.

## Introduction to use of Open-Qmin 

Open-Qmin is an open-source code base for numerical modeling of nematic liquid crystals and their topological defects at equilibrium, by relaxation of a finite-difference lattice discretization of the Landau-de Gennes free energy. It is used here for all 3D nematic simulations. Our use of Open-Qmin will be described here, but a more in depth guide can be found here: https://sussmanlab.github.io/open-Qmin/Overview.html

## Simulating experimental cell in Open-Qmin

Due to computational restrictions, we simulate a smaller representitive cell, to represent the behavior of a larger cell. 

*A justification for this could be added here*

In open-Qmin, it is necessary to replace dimensionalized quantities with non-dimensionalized counterparts. The non-dimensionalization of the system is limited by the lattice spacing, which is approximately the size of one mesogen. Here we use $\nabla x = 4.5nm$

All simulations represented here are generated with dimension (240Ã240Ã28). This represents an NLC homeotropic slab cell that is approximately 0.1 Î¼m in thickness and 1 Î¼m by 1 Î¼m wide.

```python
#More precisely 

x_units = 240
y_units = 240
z_units = 28

del_x = 4.5 * 10 ** (-9) #4.5 nm

d_x = x_units*del_x
d_y = y_units*del_x
d_z = z_units*del_x

print('lateral cell dimension (x,y) = ' + str((d_x,d_y)) + ' m')
print('simulated cell thickness (z) = ' + str(d_z) + ' m')

print('lateral cell dimension (x,y) = ' + str((d_x*10**6,d_y*10**6))  + ' um')
print('simulated cell thickness (z) = ' + str(d_z*10**6) + ' um')
```

In open-Qmin, all parameters are input via the command line. In addition, homeotropic glass walls on the top and bottom of the cell are specified as boundary objects in the openQmin.cpp file. Lateral dimensions on the other hand are periodic. 

All data, their respective generative command line scripts and changes to the cpp file are available in this repository.
